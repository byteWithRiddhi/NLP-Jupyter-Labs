```python
## A) Write a Python program to perform word-level and sentence-level tokenization on a given paragraph using NLTK or spaCy library.
## "Natural Language Processing (NLP) is a field of Artificial Intelligence. It helps computers understand human language."
```

```python
import nltk
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
nltk.download('punkt_tab')
```

```python
## WORD TOKENIZATION
```

```python
text="""Natural Language Processing (NLP) is a field of Artificial Intelligence. It helps computers understand human language."""
```

```python
token=nltk.word_tokenize(text.lower())
print("Word Tokenization:", token)
```

```python
## SENTENCE TOKENIZATION
```

```python
sentence= sent_tokenize(text)
print("Sentence Tokenization:", sentence)
```

```python
## B) Represent the given text using Bag of Words (BoW) and TF-IDF, then compare feature matrices using scikit-learn.
# Doc1: "Artificial Intelligence and Machine Learning are closely related fields."
# Doc2: "Machine Learning is a subset of Artificial Intelligence."
```

```python
import numpy as np
import pandas as pd
from collections import Counter
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
documents = [
    "Artificial Intelligence and Machine Learning are closely related fields.",
    "Machine Learning is a subset of Artificial Intelligence."
]
```

```python
## Bag of Words Representation
```

```python
bow_vectorizer = CountVectorizer()
```

```python
bow_matrix = bow_vectorizer.fit_transform(documents)
```

```python
bow_df = pd.DataFrame(
    bow_matrix.toarray(),
    columns=bow_vectorizer.get_feature_names_out()
)
```

```python
print("Bag of Words (BoW) Matrix:")
print(bow_df)
```

```python
#TF-IDF Representation
```

```python
tfidf_vectorizer = TfidfVectorizer()
```

```python
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)
```

```python
tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf_vectorizer.get_feature_names_out()
)
```

```python
print("\nTF-IDF Matrix:\n")
print(tfidf_df)
```
